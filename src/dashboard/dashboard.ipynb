## Property Values Analysis
This chart shows the distribution of property values using a log transformation.

```python
import plotly.io as pio
pio.renderers.default = "browser"
import pandas as pd
import plotly.express as px
property = pd.read_csv("/Users/samuelm/Desktop/vancouver-end-to-end-housing-insights/vancouver-end-to-end-housing-insights/data/staging/property_tax_transformed_20260415.csv")
import plotly.io as pio
pio.renderers.default = "browser"

import numpy as np

property["log_value"] = np.log1p(property["current_land_value"])

fig = px.histogram(
    property,
    x="log_value",
    nbins=50,
    title="Log Distribution of Property Values"
)

fig.show()
```

- Most properties are concentrated in a mid-range value.
- A log transformation was applied to handle extreme outliers.

## Crime by Neighbourhood

```python
crime = pd.read_csv("/Users/samuelm/Desktop/vancouver-end-to-end-housing-insights/vancouver-end-to-end-housing-insights/data/staging/crime_transformed_20260415.csv")

crime_grouped = crime.groupby("neighbourhood").size().reset_index(name="crime_count")
crime_grouped = crime_grouped.sort_values("crime_count", ascending=False)

fig = px.bar(
    crime_grouped,
    x="neighbourhood",
    y="crime_count",
    title="Crime by Neighbourhood"
)

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

- Crime incidents are concentrated in a few neighbourhoods.
- Central Business District shows the highest crime count by a large margin.

## Business Activity Across Local Areas

```python
business = pd.read_csv("/Users/samuelm/Desktop/vancouver-end-to-end-housing-insights/vancouver-end-to-end-housing-insights/data/staging/business_licences_transformed_20260415.csv")

business_grouped = business.groupby("local_area").size().reset_index(name="business_count")
business_grouped = business_grouped.sort_values("business_count", ascending=False)

fig = px.bar(
    business_grouped,
    x="local_area",
    y="business_count",
    title="Business Activity Across Local Areas"
)

fig.update_layout(
    xaxis_title="Local Area",
    yaxis_title="Number of Business Licenses",
    xaxis_tickangle=-45
)

fig.show()
```

- Business activity is highly concentrated in Downtown.
- A few local areas dominate commercial presence, while others show significantly lower activity.

## Conclusion

- Property values show a concentrated distribution after handling outliers.
- Crime is unevenly distributed, with certain neighbourhoods having significantly higher counts.
- Business activity is highly concentrated in specific local areas, especially Downtown.

These insights highlight how economic activity, safety, and property values vary across Vancouver.

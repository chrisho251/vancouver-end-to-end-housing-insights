# Exploratory Data Analysis: Local Areas
This notebook explores the Local Areas dataset.
Since this is primarily a spatial reference table mapping neighborhoods to boundaries, our analysis will focus on understanding the available regions and their central coordinates.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure visual settings
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 8)
```

```python
# Load the dataset
df = pd.read_csv('../../data/staging/local_areas_transformed_20260320.csv')

# Display the first few rows to understand the structure
display(df.head())
```

```python
# Overview of Local Areas
print(f"Total number of local areas: {len(df)}")
print("\nList of all local areas:")
print(df['name'].tolist())
```

```python
# Scatter plot of center coordinates
plt.figure()
sns.scatterplot(x='center_lon', y='center_lat', data=df, s=100, color='crimson')

# Add labels to points
for i, row in df.iterrows():
    plt.text(row['center_lon'], row['center_lat'] + 0.002, row['name'], fontsize=9, ha='center')

plt.title('Center Coordinates of Vancouver Local Areas')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
